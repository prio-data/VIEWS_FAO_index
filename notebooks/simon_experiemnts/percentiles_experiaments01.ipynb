```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_percentiles(data, percentiles):
    """
    Calculate specified percentiles for a given numpy array.
    
    Parameters:
    data (numpy array): The input array for which percentiles are calculated.
    percentiles (list of float): A list of percentiles to be calculated.
    
    Returns:
    dict: A dictionary with specified percentiles.
    """
    percentile_values = np.percentile(data, percentiles)
    return {f'{p}th Percentile': value for p, value in zip(percentiles, percentile_values)}

def show_percentile_influence_with_outlier(data, outlier_values, percentiles):
    """
    Show how the percentiles of a numpy array are influenced by adding an outlier of varying magnitude.
    
    Parameters:
    data (numpy array): The input array.
    outlier_values (list of float): A list of outlier values to be added to the data.
    percentiles (list of float): A list of percentiles to be calculated.
    
    Returns:
    None
    """
    # Calculate original percentiles without any outliers
    original_percentiles = calculate_percentiles(data, percentiles)
    
    # Initialize a dictionary to hold percentile values for each outlier
    percentile_changes = {key: [value] for key, value in original_percentiles.items()}
    
    # Calculate percentiles for each outlier value added to the data
    for outlier in outlier_values:
        modified_data = np.append(data, outlier)
        modified_percentiles = calculate_percentiles(modified_data, percentiles)
        for key, value in modified_percentiles.items():
            percentile_changes[key].append(value)
    
    # Plotting the results
    plt.figure(figsize=(14, 8))
    for key, changes in percentile_changes.items():
        plt.plot([0] + list(outlier_values), changes, label=key, marker='o')
    
    plt.xlabel('Outlier Value')
    plt.ylabel('Percentile Value')
    plt.title('Influence of Outliers on Percentiles')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = np.random.uniform(0, 100, 1000)
outlier_values = np.array([200, 500, 1000, 5000])
percentiles = [10, 20, 30, 40, 50, 60, 70, 80, 90, 95, 99, 99.99]
show_percentile_influence_with_outlier(data, outlier_values, percentiles)
```

```python
def calculate_deciles(data):
    """
    Calculate the deciles for a given numpy array.
    
    Parameters:
    data (numpy array): The input array for which deciles are calculated.
    
    Returns:
    dict: A dictionary with deciles (10th, 20th, ..., 90th percentiles).
    """
    deciles = np.percentile(data, np.arange(10, 100, 10))
    return {f'{i*10}th Percentile': decile for i, decile in enumerate(deciles, 1)}

def show_decile_influence_with_outlier(data, outlier_values):
    """
    Show how the deciles of a numpy array are influenced by adding an outlier of varying magnitude.
    
    Parameters:
    data (numpy array): The input array.
    outlier_values (list of float): A list of outlier values to be added to the data.
    
    Returns:
    None
    """
    # Calculate original deciles without any outliers
    original_deciles = calculate_deciles(data)
    
    # Initialize a dictionary to hold decile values for each outlier
    decile_changes = {f'{i*10}th Percentile': [original_deciles[f'{i*10}th Percentile']] for i in range(1, 10)}
    
    # Calculate deciles for each outlier value added to the data
    for outlier in outlier_values:
        modified_data = np.append(data, outlier)
        modified_deciles = calculate_deciles(modified_data)
        for key, value in modified_deciles.items():
            decile_changes[key].append(value)
    
    # Plotting the results
    plt.figure(figsize=(14, 8))
    for key, changes in decile_changes.items():
        plt.plot([0] + list(outlier_values), changes, label=key, marker='o')
    
    plt.xlabel('Outlier Value')
    plt.ylabel('Decile Value')
    plt.title('Influence of Outliers on Deciles')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = np.random.uniform(0, 100, 100)
outlier_values = np.array([200, 500, 1000, 5000, 10000])
show_decile_influence_with_outlier(data, outlier_values)
```

```python

def show_decile_influence_with_outlier(data, outlier_values):
    """
    Show how the deciles of a numpy array are influenced by adding an outlier of varying magnitude.
    
    Parameters:
    data (numpy array): The input array.
    outlier_values (list of float): A list of outlier values to be added to the data.
    
    Returns:
    None
    """
    # Original deciles without any outliers
    original_deciles = np.percentile(data, np.arange(10, 100, 10))
    
    # Initialize a dictionary to hold decile values for each outlier
    decile_changes = {f'{i*10}th Percentile': [] for i in range(1, 10)}
    
    # Calculate deciles for each outlier value added to the data
    for outlier in outlier_values:
        modified_data = np.append(data, outlier)
        modified_deciles = np.percentile(modified_data, np.arange(10, 100, 10))
        for i, decile in enumerate(modified_deciles):
            decile_changes[f'{(i+1)*10}th Percentile'].append(decile)
    
    # Plotting the results
    plt.figure(figsize=(14, 8))
    for i, (label, changes) in enumerate(decile_changes.items()):
        plt.plot(outlier_values, changes, label=label, marker='o')
    
    # Plot the original deciles as a reference line
    for i, decile in enumerate(original_deciles):
        plt.axhline(decile, color='grey', linestyle='--', linewidth=0.5)
        plt.text(outlier_values[-1], decile, f'Original {i*10+10}th', va='center', ha='left', fontsize=9, color='grey')
    
    plt.xlabel('Outlier Value')
    plt.ylabel('Decile Value')
    plt.title('Influence of Outlier on Deciles')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data = np.random.uniform(0, 100, 100)
outlier_values = np.array([200, 500, 1000, 5000, 10000])
show_decile_influence_with_outlier(data, outlier_values)
```
