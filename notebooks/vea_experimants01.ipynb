```python
import sys
import numpy as np
import torch

import matplotlib.pyplot as plt

sys.path.insert(0, "/home/simon/Documents/scripts/VIEWS_FAO_index/src/architectures/vea_001")

from vea_001 import VAE
```

```python
# Create a random 2D array
# np_array = np.random.rand(128, 128).astype(np.float32)
np_array = np.random.lognormal(size=(128, 128)).astype(np.float32)

# Convert to PyTorch tensor
tensor = torch.from_numpy(np_array)

# Reshape tensor to have four dimensions
tensor = tensor.unsqueeze(0).unsqueeze(0)
```

```python
# no training yet... 

vea = VAE(128, 265, 128)

reconstruction, mu, sigma = vea(tensor)
```

```python
# Visualize the original array and the pooled array
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(np_array, interpolation='nearest', cmap='viridis')
ax[0].set_title("Original")
ax[1].imshow(reconstruction.squeeze(0).squeeze(0).detach().numpy(), interpolation='nearest', cmap='viridis')
ax[1].set_title("Reconstruction")
```
